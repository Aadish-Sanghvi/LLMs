### Load a Pretrained BERT Model

```python
from transformers import AutoModel, AutoTokenizer

model_name = "bert-base-uncased"  # Lowercased English BERT model
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModel.from_pretrained(model_name)
```

### Tokenizing Text with BERT

```python
text = "Hugging Face is amazing!"
tokens = tokenizer.tokenize(text)
print(tokens)
```

```python
# Now, let's convert these tokens into input IDs:
input_ids = tokenizer.encode(text, return_tensors="pt")
print(input_ids)
```

# Fine-Tuning BERT for Sentiment Analysis

### Load a Pretrained BERT Classifier

```python
from transformers import AutoModelForSequenceClassification

model_name = "nlptown/bert-base-multilingual-uncased-sentiment"
model = AutoModelForSequenceClassification.from_pretrained(model_name)
```

### Perform Sentiment Analysis

```python
from transformers import pipeline

sentiment_analyzer = pipeline("sentiment-analysis", model=model, tokenizer=tokenizer)

# Example Reviews
reviews = ["This product is great!", "Worst experience ever!", "The food was okay, nothing special."]
results = sentiment_analyzer(reviews)

for review, result in zip(reviews, results):
    print(f"Review: {review} → Sentiment: {result}")
```
